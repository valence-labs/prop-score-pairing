```python
import wandb
import numpy as np
api = wandb.Api()
```

```python
run_id = "rkak0inz" # ps
#run_id = "lphjxkp8" # vae
run = api.run("ps-matching/Matching-Experiments/" + run_id)

# save the metrics for the run to a csv file
metrics_dataframe = run.history(samples = 9000,x_axis = "full_val_loss")
data = metrics_dataframe.loc[~np.isnan(metrics_dataframe["PS + EOT Trace Average"])].reset_index()
data["val_loss"] = metrics_dataframe.loc[~np.isnan(metrics_dataframe.full_val_loss)]["full_val_loss"].values
```

```python
import matplotlib.pyplot as plt
import matplotlib

font = {'size'   : 16,
        'family' : "sans"}

matplotlib.rc('font', **font)

val = data.val_loss.values
match_metric1 = data["PS + EOT Trace Average"] # data["PS + kNN FOSCTTM Average"] data["PS + kNN Trace Average"] data["PS + EOT Trace Average"]
match_metric2 = data["PS + kNN Trace Average"]
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.scatter(val, match_metric1*1e4, label = "PS + OT")
ax.scatter(val, match_metric2*1e4, label = "PS + SNN")
ax.set_xscale("log")
ax.xaxis.set_major_locator(matplotlib.ticker.NullLocator())
ax.xaxis.set_minor_locator(matplotlib.ticker.NullLocator())
ax.xaxis.set_major_formatter(matplotlib.ticker.NullFormatter())
ax.xaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())
ax.set_xlabel("$\longleftarrow$ Decreasing Validation Cross-Entropy", loc = "right")
ax.set_ylabel("Trace Metric ($10^{-4})$")
ax.legend()
```
